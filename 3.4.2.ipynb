```python
import requests
import pandas as pd 
import json
import openpyxl
import datetime


url = "https://jsonplaceholder.typicode.com/todos"

response = requests.get(url)

lista = []

def valores_respuesta():
    if response.status_code == 200:
        print("Respuesta con servidor aceptada")
        data = response.json()
        #print(json.dumps(data, indent=4))
     

        for user in data:
            dic = {
                "userId": user["userId"],
                "id": user["id"],
                "title": user["title"],
                "completed": user["completed"]
                }


            lista.append(dic) 
    return lista        

    
val =  valores_respuesta()
datos_web = pd.DataFrame(val)

datos_web = datos_web[datos_web["completed"]==False]
datos_web = datos_web.rename(columns = {"title": "Tarea", "id": "ID_Tarea", "userId": "ID_Usuario"})
print(datos_web.head())

today=datetime.date.today().strftime("%Y-%m-%d")
datos_web.to_excel(f"tareas_pendientes_{today}.xlsx", index=False)
print("Excel creado")



```
